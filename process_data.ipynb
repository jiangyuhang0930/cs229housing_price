```python
import pandas as pd
import numpy as np
ames = pd.read_csv('ames.csv')
melb = pd.read_csv('melb_data.csv')
```

```python
melb2 = melb[["Price", "Rooms", "Type", "Date", "Bedroom2", "Bathroom", "Car", "Landsize", "BuildingArea"]]
melb2 = melb2.dropna()
melb2
```

```python
ames2 = ames[["price", "TotRms.AbvGrd", "Bldg.Type", "Yr.Sold", "Mo.Sold", "Bedroom.AbvGr", "Full.Bath", "Half.Bath", "Garage.Cars", "Lot.Area", "area"]]
ames2["Bathroom"] = ames2["Full.Bath"] + ames2["Half.Bath"]
ames2 = ames2.drop(columns=["Full.Bath", "Half.Bath"])
ames2 = ames2.dropna()
ames2.head()
```

```python
ames2["Month_from_2000"] = (ames2["Yr.Sold"] - 2000) * 12 + ames2["Mo.Sold"]
ames2
```

```python
date1 = pd.to_datetime("1999/12/31")
melb2["Month_from_2000"] = (pd.DatetimeIndex(melb2["Date"]).year - date1.year) * 12 + (pd.DatetimeIndex(melb2["Date"]).month - date1.month)
melb2
```

```python
dic = {"1Fam":0, "2fmCon":0, "Duplex":1, "Twnhs":2, "TwnhsE":2}
ames2['Bldg.Type'] =  ames2['Bldg.Type'].map(dic)
ames2.head()
```

```python
dic2 = {'h':0,'u':1,'t':2}
melb2['Type'] = melb2['Type'].map(dic2)
melb2.head()
```

```python
ames2.head()
```

```python
ames3 = pd.get_dummies(ames2, columns=["Bldg.Type"])
ames3.head()
```

```python
ames4 = ames3.drop(columns=["Yr.Sold","Mo.Sold"])
ames_target = ames4["price"]
ames_features = ames4.drop(columns="price")
ames_features.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1,1))
ames_features = scaler.fit_transform(ames_features)
```

```python
np.savetxt('ames_features.csv',ames_features,delimiter=",")
np.savetxt('ames_target.csv',ames_target,delimiter=",")
```

```python
melb3 = pd.get_dummies(melb2, columns=["Type"])
melb3
```

```python
melb4 = melb3.drop(columns=["Date"])
melb_target = melb4["Price"]
melb_features = melb4.drop(columns="Price")
melb_features.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1,1))
melb_features = scaler.fit_transform(melb_features)
```

```python
np.savetxt('melb_features.csv',melb_features,delimiter=",")
np.savetxt('melb_target.csv',melb_target,delimiter=",")
```

```python
import my_models
from sklearn.model_selection import train_test_split
model1 = my_models.get_gradient_boost()
ames_X_train, ames_X_test, ames_y_train, ames_y_test = train_test_split(
        ames_features, ames_target, test_size=0.2)
melb_X_train, melb_X_test, melb_y_train, melb_y_test = train_test_split(
        melb_features, melb_target, test_size=0.2)
model1.fit(ames_X_train, ames_y_train)
```

```python
def evaluate_relative_error(y_pred, y):
    return np.mean(np.abs(y-y_pred) / y_pred)

def evaluate_mse(y_pred, y):
    return np.mean((y-y_pred)**2) /1e6

def train_model(model, X_train, y_train):
    model.fit(X_train, y_train)

def predict(model, X_test, y_test):
    y_pred = model.predict(X_test)
    R2 = model.score(X_test, y_test)
    rel_error = evaluate_relative_error(y_pred, y_test)
    mse = evaluate_mse(y_pred, y_test)
    return R2, rel_error, mse
```

```python
print(predict(model1, ames_X_test, ames_y_test))
```

The model preformed well on AMES data because we trained the model on that.
